```python
with open("../output/combined_text.txt", "r", encoding="utf-8") as f:
    text_sequence = f.read()

len(text_sequence)
```

```python
import sys
sys.path.append('..')
```

```python
from minbpe import BasicTokenizer

tokenizer = BasicTokenizer()
tokenizer.train(text_sequence, vocab_size=1024)
```

```python
vocab = tokenizer.vocab
vocab
```

```python
tokenizer.encode("Hi I am girish")
```

```python
tokenizer.decode([72, 258, 753, 916, 331, 104])
```

Add spcl tokens to the vocabulary. These tokens are going to be used a lot in the fine-tuning step.

```python
max_vocab_id = list(tokenizer.vocab.keys())[-1]
tokenizer.special_tokens = {
    "<|startoftext|>": max_vocab_id + 1,
    "<|separator|>": max_vocab_id + 2,
    "<|endoftext|>": max_vocab_id + 3,
    "<|unk|>": max_vocab_id + 4
}
```
